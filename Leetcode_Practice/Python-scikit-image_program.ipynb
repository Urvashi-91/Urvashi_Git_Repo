```python
img = io.imread('flowers.jpg')
io.imshow(img)
```

```python
nrows, ncols, nsize = img.shape
row, col = np.ogrid[:nrows, :ncols]
cnt_row, cnt_col = nrows / 2, ncols / 2
outer_disk_mask = ((row - cnt_row)**2 + (col - cnt_col)**2 > (nrows / 2)**2) #a^2 + b^2 = c/2^2
img[outer_disk_mask] = 0

io.imshow(img)
```
